```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
data, target = datasets.load_diabetes(return_X_y=True)
print(data.shape)
print(data[0])
```

```python
data = data[:, np.newaxis, 9]
```

```python
data_train, data_test, target_train, target_test = model_selection.train_test_split(data, target, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(data_train, target_train)
```

```python
target_pred = model.predict(data_test)
print(target_pred)
```

```python
plt.scatter(data_test, target_test, color='black')
plt.plot(data_test, target_pred, color='blue', linewidth=0.3)
plt.xlabel('Scaled BMI\'s')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
